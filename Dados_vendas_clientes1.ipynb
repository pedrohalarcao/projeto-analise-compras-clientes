```python
import requests
import pandas as pd
import io

url = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json'

try:
    response = requests.get(url)
    response.raise_for_status()
    dados = pd.read_json(io.StringIO(response.text))
    display(dados.head())
except requests.exceptions.RequestException as e:
    print(f"Error fetching the data: {e}")
except ValueError as e:
    print(f"Error parsing JSON data: {e}")
```

```python
dados = pd.json_normalize(dados['dados_vendas'])
dados.head()
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(['Cliente', 'Valor da compra'])
dados.head()
```

```python
dados.reset_index(inplace = True,drop=True)
dados.head()
```

```python
dados.info()
```

```python
dados['Valor da compra'] = dados['Valor da compra'].apply(lambda x: x.replace('R','').replace('$','').replace(',','').strip())
```

```python
import numpy as np
```

```python
dados['Valor da compra'] = dados['Valor da compra'].astype(np.float64)
dados.head()
```

```python
dados.info()
```

```python
dados['Cliente'] = dados['Cliente'].str.lower()
```

```python
dados
```

```python
dados['Cliente'] = dados['Cliente'].str.replace(r'[^a-z\s]',' ',regex=True)
dados.head()
```

```python
dados['Data de venda'] = pd.to_datetime(dados['Data de venda'])
dados.head()
```

```python
dados.info()
```

# Objetivo
Criar um subset para visualizar qual foi o cliente que mais comprou no periodo de 5 dias.

## Definir o período de 5 dias

### Subtask:
Determinar as datas de início e fim para o período de 5 dias de interesse.


**Raciocínio**:
Calcular as datas de início e fim para o período de 5 dias com base na coluna 'Data de venda' no dataframe. A data de início será a data mínima e a data de fim será a data mínima mais 4 dias (para incluir a data de início e formar um período de 5 dias).



```python
data_inicial = dados['Data de venda'].min()
data_final = data_inicial + pd.Timedelta(days=4)
print(f"Data inicial: {data_inicial}")
print(f"Data final: {data_final}")
```

## Filtrar os dados

### Subtask:
Criar um novo DataFrame contendo apenas os dados de vendas dentro do período de 5 dias definido.


**Raciocínio**:
Filtrar o dataframe com base nas datas de início e fim para criar um novo dataframe para o período de 5 dias.






```python
dados_periodo = dados[(dados['Data de venda'] >= data_inicial) & (dados['Data de venda'] <= data_final)].copy()
dados_periodo.head()
```

## Agrupar e agregar

### Subtask:
Agrupar os dados filtrados por 'Cliente' e somar o 'Valor da compra' para cada cliente.


**Raciocínio**:
Agrupar os dados filtrados por 'Cliente' e somar o 'Valor da compra' para cada cliente para encontrar o valor total de compra por cliente dentro do período especificado de 5 dias.



```python
total_compra_por_cliente = dados_periodo.groupby('Cliente')['Valor da compra'].sum()
display(total_compra_por_cliente)
```

## Identificar o cliente com maior compra

### Subtask:
Encontrar o cliente com o maior valor total de compra a partir dos dados agregados.


**Raciocínio**:
Encontrar o cliente com o valor total máximo de compra usando idxmax() e obter o valor total de compra dele.



```python
cliente_maior_compra = total_compra_por_cliente.idxmax()
valor_maior_compra = total_compra_por_cliente[cliente_maior_compra]
print(f"O cliente que mais comprou no período de 5 dias é: {cliente_maior_compra}")
print(f"O valor total da compra deste cliente é: {valor_maior_compra}")
```

## Exibir o resultado

### Subtask:
Mostrar o nome do cliente que comprou mais e o valor total de suas compras dentro do período de 5 dias.


**Raciocínio**:
Mostrar o cliente com a maior compra e o valor total de compra dele.



```python
print(f"O nome do cliente que comprou mais no período de 5 dias é: {cliente_maior_compra}")
print(f"O valor total das compras deste cliente neste período é: {valor_maior_compra}")
```
